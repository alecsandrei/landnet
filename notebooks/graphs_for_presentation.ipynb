```python
from __future__ import annotations

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from landnet.config import MODELS_DIR
```

```python
metrics = MODELS_DIR / 'convnext_100x100' / 'results_summary.csv'
df = pd.read_csv(metrics).iloc[:, 1:].set_index('variable')
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
sns.barplot(df, x='sensitivity', y='variable', orient='y', ax=ax)

plt.tight_layout()
plt.savefig(metrics.parent / 'metrics.png', dpi=300)
```
